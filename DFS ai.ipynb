```python
# DFS Implementation for Travelling Salesman Problem (Arad to Bucharest)

# 1. Representing the Graph as an Adjacency List
# Each key is a city, and its value is a list of tuples (neighbor, distance)
graph = {
    'Arad': [('Zerind', 75), ('Sibiu', 140), ('Timisoara', 118)],
    'Zerind': [('Arad', 75), ('Oradea', 71)],
    'Oradea': [('Zerind', 71), ('Sibiu', 151)],
    'Sibiu': [('Arad', 140), ('Oradea', 151), ('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Timisoara': [('Arad', 118), ('Lugoj', 111)],
    'Lugoj': [('Timisoara', 111), ('Mehadia', 70)],
    'Mehadia': [('Lugoj', 70), ('Drobeta', 75)],
    'Drobeta': [('Mehadia', 75), ('Craiova', 120)],
    'Craiova': [('Drobeta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Rimnicu Vilcea': [('Sibiu', 80), ('Craiova', 146), ('Pitesti', 97)],
    'Fagaras': [('Sibiu', 99), ('Bucharest', 211)],
    'Pitesti': [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Bucharest': [('Fagaras', 211), ('Pitesti', 101), ('Giurgiu', 90), ('Urziceni', 85)],
    'Giurgiu': [('Bucharest', 90)],
    'Urziceni': [('Bucharest', 85), ('Vaslui', 142), ('Hirsova', 98)],
    'Hirsova': [('Urziceni', 98), ('Eforie', 86)],
    'Eforie': [('Hirsova', 86)],
    'Vaslui': [('Urziceni', 142), ('Iasi', 92)],
    'Iasi': [('Vaslui', 92), ('Neamt', 87)],
    'Neamt': [('Iasi', 87)]
}

def dfs_path(graph, start, goal):
    # Stack stores (current_node, path_taken, total_distance)
    stack = [(start, [start], 0)]
    visited = set()

    while stack:
        (vertex, path, distance) = stack.pop()

        if vertex in visited:
            continue

        # Check if we reached the goal
        if vertex == goal:
            return path, distance

        visited.add(vertex)

        # Explore neighbors
        for neighbor, weight in graph.get(vertex, []):
            if neighbor not in visited:
                new_path = path + [neighbor]
                new_distance = distance + weight
                stack.append((neighbor, new_path, new_distance))

    return None, 0

# Execute the search
start_node = 'Arad'
goal_node = 'Bucharest'
result_path, total_dist = dfs_path(graph, start_node, goal_node)

# Output Results
print("--- Depth First Search Traversal ---")
if result_path:
    print(f"Path found: {' -> '.join(result_path)}")
    print(f"Total Distance: {total_dist} km")
else:
    print("Goal not reachable.")
```

```python
import collections

# The Graph from the TSP Figure
graph = {
    'Arad': [('Zerind', 75), ('Sibiu', 140), ('Timisoara', 118)],
    'Zerind': [('Arad', 75), ('Oradea', 71)],
    'Oradea': [('Zerind', 71), ('Sibiu', 151)],
    'Sibiu': [('Arad', 140), ('Oradea', 151), ('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Timisoara': [('Arad', 118), ('Lugoj', 111)],
    'Lugoj': [('Timisoara', 111), ('Mehadia', 70)],
    'Mehadia': [('Lugoj', 70), ('Drobeta', 75)],
    'Drobeta': [('Mehadia', 75), ('Craiova', 120)],
    'Craiova': [('Drobeta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Rimnicu Vilcea': [('Sibiu', 80), ('Craiova', 146), ('Pitesti', 97)],
    'Fagaras': [('Sibiu', 99), ('Bucharest', 211)],
    'Pitesti': [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Bucharest': [('Fagaras', 211), ('Pitesti', 101), ('Giurgiu', 90), ('Urziceni', 85)],
    'Giurgiu': [('Bucharest', 90)],
    'Urziceni': [('Bucharest', 85), ('Vaslui', 142), ('Hirsova', 98)],
    'Hirsova': [('Urziceni', 98), ('Eforie', 86)],
    'Eforie': [('Hirsova', 86)],
    'Vaslui': [('Urziceni', 142), ('Iasi', 92)],
    'Iasi': [('Vaslui', 92), ('Neamt', 87)],
    'Neamt': [('Iasi', 87)]
}

def run_bfs(start, goal):
    queue = collections.deque([[start]])
    visited = set()
    nodes_visited_count = 0

    while queue:
        path = queue.popleft()
        node = path[-1]
        nodes_visited_count += 1

        if node == goal:
            return nodes_visited_count, path

        if node not in visited:
            for neighbor, weight in graph.get(node, []):
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)
            visited.add(node)
    return nodes_visited_count, None

def run_dfs(start, goal):
    stack = [[start]]
    visited = set()
    nodes_visited_count = 0

    while stack:
        path = stack.pop()
        node = path[-1]
        nodes_visited_count += 1

        if node == goal:
            return nodes_visited_count, path

        if node not in visited:
            # Reversing neighbors to maintain same order as BFS for comparison
            neighbors = graph.get(node, [])
            for neighbor, weight in reversed(neighbors):
                new_path = list(path)
                new_path.append(neighbor)
                stack.append(new_path)
            visited.add(node)
    return nodes_visited_count, None

def compare(start, goal):
    bfs_count, bfs_path = run_bfs(start, goal)
    dfs_count, dfs_path = run_dfs(start, goal)
    print(f"Start: {start} | Goal: {goal}")
    print(f"BFS Visited: {bfs_count} nodes | Path: {' -> '.join(bfs_path)}")
    print(f"DFS Visited: {dfs_count} nodes | Path: {' -> '.join(dfs_path)}")
    print("-" * 30)

# Scenario 1: BFS is smaller than DFS
# Goal is shallow (neighbor)
print("Scenario 1: BFS < DFS")
compare('Arad', 'Zerind')

# Scenario 2: BFS is larger than DFS
# Goal is deep in the first branch DFS explores
print("Scenario 2: BFS > DFS")
compare('Arad', 'Lugoj')
```

```python
# Activity 1: Recursive Depth-First Search Implementation

# 1. Representing the Graph using an Adjacency List (Dictionary)
# Keys are nodes, and values are lists of connected (neighbor) nodes
graph = {
    '5' : ['3', '7'],
    '3' : ['2', '4'],
    '7' : ['8'],
    '2' : [],
    '4' : ['8'],
    '8' : []
}

# 2. Set to keep track of visited nodes to avoid infinite loops
visited = set()

def dfs(visited, graph, node):
    """
    Function to perform DFS starting from a specific node.
    - visited: set containing nodes already seen
    - graph: the adjacency list
    - node: the current node being explored
    """
    if node not in visited:
        # Step 2: Mark the node as visited and print it
        print(node, end=" ")
        visited.add(node)

        # Step 3: Explore all adjacent nodes (neighbors)
        for neighbour in graph[node]:
            # Recursive call to dive deeper into the branch
            dfs(visited, graph, neighbour)

# --- Driver Code ---
print("Following is the Depth-First Search traversal:")
# Starting the search from node '5'
dfs(visited, graph, '5')
```

```python
from collections import defaultdict

# This class represents a directed graph using adjacency list representation
class Graph:
    # Constructor
    def __init__(self):
        # default dictionary to store graph
        self.graph = defaultdict(list)

    # function to add an edge to graph
    def addEdge(self, u, v):
        self.graph[u].append(v)

    # A function used by DFS (Recursive Helper)
    def DFSUtil(self, v, visited):
        # Mark the current node as visited and print it
        visited[v] = True
        print(v, end=" ")

        # Recur for all the vertices adjacent to this vertex
        for i in self.graph[v]:
            if visited[i] == False:
                self.DFSUtil(i, visited)

    # The function to do DFS traversal.
    # It ensures all vertices are visited even if the graph is disconnected.
    def DFS(self):
        # Identify the number of unique vertices
        # We find the max index to handle cases where len(graph) might be misleading
        if not self.graph:
            return

        nodes = set(self.graph.keys())
        for values in self.graph.values():
            nodes.update(values)

        V = len(nodes)

        # Mark all the vertices as not visited
        visited = [False] * V

        # Call the recursive helper function for every vertex
        # that hasn't been visited yet
        for i in range(V):
            if visited[i] == False:
                self.DFSUtil(i, visited)

# --- Driver code ---
# Create a graph as shown in your lab diagram
g = Graph()
g.addEdge(0, 1)
g.addEdge(0, 2)
g.addEdge(1, 2)
g.addEdge(2, 0)
g.addEdge(2, 3)
g.addEdge(3, 3)

print("Following is Depth First Traversal:")
g.DFS()
```
