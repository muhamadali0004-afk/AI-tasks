```python

romania_map = {
    'Arad': [('Sibiu', 140), ('Timisoara', 118), ('Zerind', 75)],
    'Sibiu': [('Arad', 140), ('Fagaras', 99), ('Rimnicu Vilcea', 80)],
    'Timisoara': [('Arad', 118), ('Lugoj', 111)],
    'Zerind': [('Arad', 75), ('Oradea', 71)],
    'Fagaras': [('Sibiu', 99), ('Bucharest', 211)],
    'Rimnicu Vilcea': [('Sibiu', 80), ('Pitesti', 97), ('Craiova', 146)],
    'Lugoj': [('Timisoara', 111), ('Mehadia', 70)],
    'Mehadia': [('Lugoj', 70), ('Drobeta', 75)],
    'Drobeta': [('Mehadia', 75), ('Craiova', 120)],
    'Craiova': [('Drobeta', 120), ('Rimnicu Vilcea', 146), ('Pitesti', 138)],
    'Pitesti': [('Rimnicu Vilcea', 97), ('Craiova', 138), ('Bucharest', 101)],
    'Bucharest': [('Fagaras', 211), ('Pitesti', 101)]
}

from collections import deque

def bfs_romania(start, goal):
    queue = deque([(start, 0, [start])])
    visited = set()

    while queue:
        node, cost, path = queue.popleft()
        if node == goal:
            print("Path:", " -> ".join(path))
            print("Total Distance:", cost)
            return
        if node not in visited:
            visited.add(node)
            for neighbour, dist in romania_map.get(node, []):
                if neighbour not in visited:
                    queue.append((neighbour, cost + dist, path + [neighbour]))

bfs_romania('Arad', 'Bucharest')
```
