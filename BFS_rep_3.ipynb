```python
# Activity 3: BFS on Graph (A-G)
graph = {
    'A': ['B', 'E'],
    'B': ['A', 'C', 'D'],
    'C': ['B', 'F'],
    'D': ['B', 'G'],
    'E': ['A', 'F'],
    'F': ['C', 'E', 'G'],
    'G': ['D', 'F']
}

visited = []
queue = []

def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)

    while queue:
        m = queue.pop(0)
        print(m, end=" ")

        for neighbour in graph[m]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)

print("BFS starting from A:")
bfs(visited, graph, 'A')
```
